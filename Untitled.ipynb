```python
import pandas as pd
import os
import csv
import datetime as dt
```

```python
f_path=os.path.join("NBA_PER.csv")
stats_df=pd.read_csv(f_path)
del stats_df["Unnamed: 0.1"]
del stats_df["Unnamed: 0"]
stats_df["SEASON_ID"].astype("float64")
stats_df=stats_df.loc[stats_df["SEASON_ID"]<=2017]
stats_df
```

```python
f_path=os.path.join("Player - Salaries per Year (1990 - 2017).csv")
Salaries_df=pd.read_csv(f_path)
Salaries_df
```

```python
merge_df=stats_df.merge(Salaries_df, left_on=['full_name','SEASON_ID'], right_on=['Player Name','Season End'])
merge_df.to_csv('NBA_Players_Stats_Salaries.csv')
# merge_df.keys()
```
