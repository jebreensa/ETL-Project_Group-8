```python
import pandas as pd
import os
import csv
from nba_api.stats.endpoints import commonplayerinfo
from nba_api.stats.static import players
from nba_api.stats.static import teams
from nba_api.stats.endpoints import playercareerstats
from time import sleep
```

```python
# get_players returns a list of dictionaries, each representing a player.
nba_players = players.get_players()
print('Number of players fetched: {}'.format(len(nba_players)))
df=nba_players

# career = playercareerstats.PlayerCareerStats(player_id='203076')
# career.get_data_frames()[0]





```

```python
df = pd.DataFrame(df)
df
player_id= df['id'].tolist()

stats_df = playercareerstats.PlayerCareerStats(player_id=str(player_id[0])).get_data_frames()[0]
stats_df.head()
```

```python
for i in player_id[player_id.index(stats_df['PLAYER_ID'].tolist()[-1])+1::]:
    print(i, end='\r')
    try:
        stats_df = stats_df.append(playercareerstats.PlayerCareerStats(player_id=str(i)).get_data_frames()[0], ignore_index=True)
    except Exception as e:
        print('it broke')
        break
```

```python
# playercareerstats.PlayerCareerStats(player_id=str(player_id[0])).get_data_frames()[0]
merge_df=stats_df.merge(df, left_on='PLAYER_ID', right_on='id')
merge_df.to_csv('NBA_Players_Stats.csv')
```
