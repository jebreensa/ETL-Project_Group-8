```python
import pandas as pd
import scipy.stats as st
import numpy as np
import os
```

```python
file_to_load = ("NBA_Players_Stats_Updated.csv")
# Read Purchasing File and store into Pandas data frame
NBA_Pstats = pd.read_csv(file_to_load)
```

```python
pd.set_option('display.max_columns', None)
NBA_Pstats.tail()
               
```

```python
print(len(NBA_Pstats.columns))
```

```python
#calculating players efficiency non-adjusted (estimation)

FGM_P = NBA_Pstats['FGM'] * 85.91
STL_P = NBA_Pstats['STL'] * 53.897
FG3M_P = NBA_Pstats['FG3M'] * 51.757
FTM_P = NBA_Pstats['FTM'] * 46.845
BLK_P = NBA_Pstats['BLK'] * 39.19
OREB_P = NBA_Pstats['OREB'] * 39.19
AST_P = NBA_Pstats['AST'] *34.677
DREB_P = NBA_Pstats['DREB'] * 14.707
PF_P = NBA_Pstats['PF'] * 17.174 
FTmiss_P = ((NBA_Pstats['FTA'])-(NBA_Pstats['FTM']))* 20.091
FGmiss_P = ((NBA_Pstats['FGA'])-(NBA_Pstats['FGM']))* 39.19
TO_P = NBA_Pstats['TOV'] * 53.897
MIN_P = NBA_Pstats['MIN']

Per = (FGM_P + STL_P + FG3M_P + FTM_P + BLK_P + OREB_P + AST_P + DREB_P - PF_P - FTmiss_P - FGmiss_P - TO_P)*(1/MIN_P)
```

```python
Per
```

```python
NBA_Pstats.insert(36, 'PER', Per)

```

```python
NBA_Pstats.tail()
```

```python
NBA_new =NBA_Pstats.rename(columns={"Ave.": "CPI_AVE"})
```

```python
file_to_load = ("NBA_Players_Stats_Salaries.csv")
NBA_salary = pd.read_csv(file_to_load)

NBA_salary.tail(30)
```

```python
index = NBA_salary.index
number_of_rows = len(index)
number_of_rows
```

```python
Over_stats = NBA_salary[['PLAYER_ID','CPI_AVE', 'PER', ' Salary in $ ']]
Over_stats.head()
```

```python
Over_stats[' Salary in $ '] = Over_stats[' Salary in $ '].replace({'\$': '', ',': ''}, regex=True).astype(float)
```

```python
Over_stats.dtypes
```

```python
# Over_stats['PER'] =Over_stats['PER'].astype(str).astype(float)
Over_stats['PER'] = pd.to_numeric(Over_stats['PER'], errors='coerce')
```

```python

Over_Player_stats =Over_stats.rename(columns={' Salary in $ ': 'salary'})
Over_Player_stats
```

```python
#calculating player undepaid and overpaid
Sal_CPI_Norm = Over_Player_stats['salary']/Over_Player_stats['CPI_AVE']

Adj_salary = 15000000*(Over_Player_stats['CPI_AVE']/245.12)

salary_ratio = Adj_salary/Over_Player_stats['CPI_AVE']

Sal_Ave_player_norm = Sal_CPI_Norm/salary_ratio

Per_Norm_Aver = Over_Player_stats['PER']/15

Norm_PER_Sal = Sal_Ave_player_norm/Per_Norm_Aver

over_under= Norm_PER_Sal-1

over_under
```

```python
Over_under= pd.DataFrame(over_under)
Over_under.tail(30)
```
