# Put pandas dataframes into postgres


```python
import pandas as pd
from sqlalchemy import create_engine
import os
```

### Read *NBA_Player_Stats* CSV

```python
stats = os.path.join("NBA_PLayer_Evaluation.csv")
stats_df = pd.read_csv(stats)
stats_df.head()
```

```python
#remove unnecessary columns
del stats_df['Unnamed: 0']
del stats_df['Season Start']
del stats_df['TEAM_ID']
del stats_df['TEAM_ABBREVIATION']
del stats_df['Season End']
del stats_df['Team']
del stats_df['id']
del stats_df['full_name']
del stats_df['first_name']
del stats_df['last_name']
del stats_df['LEAGUE_ID']
stats_df.head()
```

```python
# rearrange columns
stats_df_rearrange = stats_df[['PLAYER_ID', 'Player Name', 'PLAYER_AGE', 'Full Team Name', 'is_active', 'SEASON_ID', 'GP', 'GS', 'MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA', 'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', ' Salary in $ ', 'PER', 'CPI_AVE', 'Register Value', 'Player_Evaluation']]
stats_df_rearrange.head()
```

### Transform NBA stats DataFrame

```python
# Rename the column headers to lowercase
stats_df_rearrange.columns= stats_df_rearrange.columns.str.lower()

stats_df_rearrange.head()
```

```python
##rename columns to remove spaces
## comment out after running!
stats_df_rearrange.rename(columns={"player name": "player_name", "full team name": "full_team_name", " salary in $ ": "salary", "register value": "register_value"}, inplace = True)
stats_df_rearrange.head()

#remove dollar sign
stats_df_rearrange['salary'] = stats_df_rearrange['salary'].str.replace(',', '')
stats_df_rearrange['salary'] = stats_df_rearrange['salary'].str.replace('$', 
                                                                        
                                                                        
for col in stats_df_rearrange.columns: 
    print(col) 
stats_df_rearrange.head()
```

```python
#remove rows with error values


for index, row in stats_df_rearrange.iterrows():
    if row['per'] == '#NAME?':
        stats_df_rearrange.drop(index, inplace=True)
        
for index, row in stats_df_rearrange.iterrows():
    if row['per'] == 'inf':
        stats_df_rearrange.drop(index, inplace=True)
        

    
```
